```python
import pandas as pd
```

```python
tables = pd.read_html('http://www.cis.es/cis/export/sites/default/-Archivos/Indicadores/documentos_html/TresProblemas.html')
```

```python
# We have 2 tables, the first is for a better presentation. We use the second
df = tables[1]
df
```

```python
# The row 1 work as name columns
df.columns = df.iloc[0]
df
```

```python
# Serie 1 will work as index. Then we will delete it
df1 = df.set_index(df.iloc[:,0].values)
df1 = df1.drop(df1.columns[0], axis=1)
df1
```

```python
# The first three rows it will deleted.
df2 = df1.iloc[2:]
df2
```

```python
# we have 115 rows. We will delete intermediate empty rows
df3 = df2.dropna(how='all') 
# We comprobate the rows deleted are the waited
assert len(df2.index)/2 + 1 == len(df3.index)
```

```python
# We need to convert values from str unicode to float but the dataframe has values not numeric as dots '.'
# Replace dots by 0.0
df_clean = df3.replace('.', 0.0)
# Convert to numericvalues
df_clean = df_clean.apply(pd.to_numeric)
df_clean
```

```python
# We store the last row with the num de participants in another serie
poblation_serie = pd.Series(df_clean.iloc[-1],index=df_clean.columns)
# We remove the last row
df_clean = df_clean[:-1]
poblation_serie.head(10)
```

```python
# We get the last date, it`s said, the first column, for our calculus
last_month = df_clean.ix[:,0]
top_last_month = last_month.sort_values(ascending=False).head(30)
top_last_month
```

```python
# We going to get those who value is equal o greater than 4, since from this point the variations bewteen topics are small
list_topics = top_last_month[top_last_month > 4].index
df_topics = df_clean.ix[list_topics]
df_topics
```
